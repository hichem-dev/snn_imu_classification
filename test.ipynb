```python
import os
import pandas as pd

# 1️⃣ Load ALL data

data = []
labels = []

for label in os.listdir("dataset"):
    folder = os.path.join("dataset", label) if os.path.isdir(os.path.join("dataset", label)) else None

    if folder:
        for file in os.listdir(folder):
            if file.lower().endswith(".csv"):
                print(f"Loading {file} from {label}...")
                df = pd.read_csv(os.path.join(folder, file))
                data.append(df)
                labels.append(label)
labels
data
```

```python
window_size = 50
```

```python
def create_windows(df, window_size):
    windows = []
    for i in range(0, len(df) - window_size, window_size):
        windows.append(df.iloc[i:i+window_size])
    return windows
```

```python
def normalize(df):
    return df / df.abs().max()
```
